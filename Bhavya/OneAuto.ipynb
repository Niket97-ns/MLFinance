```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tickers = ['AAPL', 'MSFT', 'GOOG']
data_source = 'yahoo'
start_date = '2017-01-01'
end_date = '2017-09-28'

panel_data = data.DataReader(tickers, data_source, start_date, end_date)
```

```python
close = panel_data.loc['Close']
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')
close = close.reindex(all_weekdays)
hand = close.plot(title="Close Prices", fontsize=12)
hand.set_xlabel("Date")
hand.set_ylabel("Price")
plt.show()
```

```python
close2 = panel_data.loc['Close']
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')
close2 = close2.reindex(all_weekdays)
close2 /= close2.iloc[1,:]
hand2 = close2.plot(title="Comparative Growth(Based on Close Price)", fontsize=12)
hand2.set_xlabel("Date")
hand2.set_ylabel("Price")
plt.show()
```
