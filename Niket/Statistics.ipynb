```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tickers = ['AAPL', 'MSFT', 'GOOG']
data_source = 'yahoo'
start_date = '2017-01-01'
end_date = '2017-09-28'

panel_data = data.DataReader(tickers, data_source, start_date, end_date)

close = panel_data.loc['Close']
all_weekdays = pd.date_range(start=start_date, end=end_date, freq='B')
close = close.reindex(all_weekdays)
```

```python
plt.figure(figsize=(20,10))
ax = close['GOOG'].plot(title="Google Rolling Mean", fontsize=12, label='GOOG')
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.legend(loc='upper left')

rollingMeanGoogle = pd.rolling_mean(close['GOOG'], window=20)
rollingMeanGoogle.plot()

rollingStdGoogle = pd.rolling_std(close['GOOG'], window=20)

upperBand = rollingMeanGoogle + rollingStdGoogle*2
lowerBand = rollingMeanGoogle - rollingStdGoogle*2

upperBand.plot()
lowerBand.plot()

plt.show()
```
